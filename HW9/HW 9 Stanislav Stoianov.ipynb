### Анализ акций компании Volkswagen

```python
from itertools import tee
import pandas as pd
import plotly.express as px
from pandas import plotting
```

```python
vw = pd.read_csv(r'data/VOW3.DE.csv', encoding='utf-8', sep=',')
vw.head(1)
```

```python
px.line(vw, y='Close', x='Date', title='Цена на акции VW')
```

```python
px.histogram(vw, x='Open', title='Распределение цены VW на открытии')
```

```python
def pairwise(iterable):
    a, b = tee(iterable)
    next(b, None)
    return zip(a, b)

val = []
for v, w in pairwise(list(vw.Open.values)):
    val.append((v - w) / v)
px.histogram(val, title='Распределение доходности VW')
```

```python
plotting.scatter_matrix(vw[['Date','Open', 'Close']])
```

```python
px.line(vw, y='Volume', x='Date', title='Величина торгов акциями VW')
```
