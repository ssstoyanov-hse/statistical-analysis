##### Анализ акций Tesla и Moderna

```python
from itertools import tee
import pandas as pd
import plotly.express as px

tesla_stock = pd.read_csv(r'hw 7 data/TSLA.csv', encoding='utf-8', sep=',')
moderna_stock = pd.read_csv(r'hw 7 data/MRNA.csv', encoding='utf-8', sep=',')
```

```python
px.line(tesla_stock, y='Open', x='Date', title='Цена на акции TESLA')
```

```python
px.histogram(tesla_stock, x='Open', title='Распределение цены TESLA на открытии')
```

```python
def pairwise(iterable):
    a, b = tee(iterable)
    next(b, None)
    return zip(a, b)

val = []
for v, w in pairwise(list(tesla_stock.Open.values)):
    val.append((v - w) / v)
px.histogram(val, title='Распределение доходности TESLA')
```

```python
tesla_stock['volatility'] = pow(tesla_stock['High'] - tesla_stock['Low'], 2)
px.histogram(tesla_stock, x='volatility', title='Распределение волатильности TESLA')
```

```python
px.line(tesla_stock.volatility)
```

```python
print(f"Медиана: {tesla_stock.Open.median()}")
print(f"Среднее: {tesla_stock.Open.mean()}")
print(f"Дисперсия: {tesla_stock.Open.var()}")
```
